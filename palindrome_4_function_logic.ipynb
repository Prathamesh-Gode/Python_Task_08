# Palindrome 4 Function Logic

```python
sent1 = 'A man, a plan, a canal: Panama'
print(sent1)
```

```python
def clean_string(s):
    cleaned = ''
    for ch in s:
        if ch.isalnum():
            cleaned += ch.lower()
    return cleaned

cleaned_sent = clean_string(sent1)
print(cleaned_sent)
```

## Function 1: Palindrome Using String Slicing

```python
def palindrome_slice(s):
    s = clean_string(s)
    return s == s[::-1]

print(palindrome_slice(sent1))
```

## Function 2: Palindrome Using Loop 

```python
def palindrome_loop(s):
    s = clean_string(s)
    left = 0
    right = len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

print(palindrome_loop(sent1))
```

## Function 3: Palindrome Using reversed

```python
def palindrome_reversed(s):
    s = clean_string(s)
    rev = ''.join(reversed(s))
    return s == rev

print(palindrome_reversed(sent1))
```

## Function 4: Palindrome Using Recursion

```python
def palindrome_recursive(s):
    s = clean_string(s)
    def check(sub):
        if len(sub) <= 1:
            return True
        if sub[0] != sub[-1]:
            return False
        return check(sub[1:-1])
    return check(s)

print(palindrome_recursive(sent1))
```
